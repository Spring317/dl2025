```python
def gradient_descend_2d(old_w0, old_w1, x, y, lr, iterations = 1000): 
    loss = []
    # xs = []
    old_f = None
    count = 0
    w0 = old_w0
    w1 = old_w1
    while True:
        f_prime = w1 * x + w0 - y
        w0 = w0 - lr*f_prime
        w1 = w1 - lr*f_prime  
        f = 1/2 * (w1* x + w0 - y)**2
        loss.append(f)
        # xs.append([w0, w1])
        print(f"time: {count} w0: {w0}, w1: {w1}, f(x): {f}")
        
        if (old_f is not None and abs(old_f - f) < 1):
            break
        
        old_f = f
        count += 1    
    return loss, w0, w1 
```
